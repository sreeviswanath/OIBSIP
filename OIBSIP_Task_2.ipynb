## **Load and Preview the Dataset**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data_path = '/content/Unemployment in India.csv'
df = pd.read_csv(data_path)
df
```

## **Display the first few rows of the dataset**

```python
print("First 5 rows of the dataset:")
print(df.head())
```

## **Check for Missing Values and Basic Statistics**

```python
print("\nMissing values in each column:")
print(df.isnull().sum())
```

## **Display basic statistics**

```python
print("\nBasic statistics of numerical columns:")
print(df.describe())
```

## **Data Cleaning**

```python
df.columns = df.columns.str.strip()  # Remove any leading/trailing spaces in column names
```

## **Check the unique regions**

```python
if 'Region' in df.columns:
    print("\nUnique regions in the dataset:")
    print(df['Region'].unique())
```

## **Convert the 'Date' column to datetime**

```python
df.isetitem(df.columns.get_loc('Date'), pd.to_datetime(df['Date']))
df
```

```python
# Check the structure of the dataset after cleaning
print("\nDataset info after cleaning:")
print(df.info())
```

## **adding month number and month name and year columns for analysis**

```python
#adding the month_number
df.loc[:, 'Month_num'] = pd.to_datetime(df['Date']).dt.month
# adding the month name
df.loc[:,'Month_name'] = pd.to_datetime(df['Date']).dt.month_name()
# adding the Year
df.loc[:,'Year'] = pd.to_datetime(df['Date']).dt.year
df
```

## **Unemployment Rate based on states**

```python
data1 = df[['Region','Estimated Unemployment Rate (%)']].groupby('Region').sum().sort_values(by='Estimated Unemployment Rate (%)',ascending=False)
data1
```

```python
import plotly.express as px
plot = px.pie(df,values='Estimated Unemployment Rate (%)',names = df['Region'],title="UNEMPLOYMENT RATE IN YEAR 2019-2020(state wise)")
plot.show()
```

## **Unemployment Rate based on months**

```python
data2 = df[['Month_name','Estimated Unemployment Rate (%)']].groupby('Month_name').sum().sort_values(by='Estimated Unemployment Rate (%)',ascending=False)
data2
```

```python
import plotly.express as px
plot = px.pie(df,values='Estimated Unemployment Rate (%)',names= df['Month_name'],title="UNEMPLOYMENT RATE IN YEAR 2019-2020(Monthwise)")
plot.show()
```

## **Correlation matrix**

```python
cor = df.corr(numeric_only = True)
sns.heatmap(cor,annot=True)
```

## **Unemployment according to Area**

```python
sns.histplot(x='Estimated Unemployment Rate (%)', data=df, kde=True, hue='Area')
plt.title('Unemployment according to Area')
plt.xlabel('Unemployment Rate')
plt.show()
```

## **Unemployment according to Date**

```python
sns.lineplot(y='Estimated Unemployment Rate (%)', x='Date', data=df)
plt.title('Unemployment according to Date')
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.ylabel('Unemployment Rate')
plt.show()
```

## **Labour Participation according to Area**

```python
sns.histplot(x='Estimated Labour Participation Rate (%)', data=df, kde=True, hue='Area')
plt.title('Labour Participation according to Area')
plt.xlabel('Labour Participation Rate')
plt.show()
```

## **Labour Participation according to Date**

```python
sns.lineplot(y='Estimated Labour Participation Rate (%)', x='Date', data=df)
plt.title('Labour Participation according to Date')
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.ylabel('Labour Participation Rate')
plt.show()
```

## **Basic inferences**

```python
print("State with highest Unemployment Rate:\t",df['Region'].value_counts().idxmax())
print("State with Lowest Unemployment Rate:\t",df['Region'].value_counts().idxmin())
print("Area with the Highest Unemployment Rate:",df['Area'].value_counts().idxmax())
print("Area with the Lowest Unemployment Rate:\t",df['Area'].value_counts().idxmin())
print("Month with Highest Unemployment Rate:\t",df['Month_name'].value_counts().idxmax())
print("Month with Lowest Unemployment Rate:\t",df['Month_name'].value_counts().idxmin())
print("Year with Highest Unemployment Rate(2019 or 2020):",df['Year'].value_counts().idxmax())
```
